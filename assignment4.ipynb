```python
#1.Load the patient dataset and show summary with info().
```

```python
df=pd.read_csv('Patient_Data.csv')
df1=pd.read_csv('Billing_Data.csv')
```

```python
df.info()
```

```python
#2.	Select only the columns relevant for billing: ['PatientID', 'Department', 'Doctor', 'BillAmount']
```

```python
pd.DataFrame(df,columns=['PatientID', 'Department', 'Doctor', 'BillAmount'])
```

```python
#Drop administrative columns like ['ReceptionistID', 'CheckInTime'].
```

```python
df.drop(['ReceptionistID', 'CheckInTime'],axis=1,inplace=True)
```

```python
df
```

```python
#4. Use groupby to find total bill amount per department.
```

```python
print((df.groupby('Department')['BillAmount'].sum()))
```

```python
#5. Remove duplicate patient records based on PatientID.
```

```python
df.drop_duplicates(['PatientID'],inplace=True)
```

```python
df
```

```python
#6. Fill missing BillAmount values with the mean bill amount.
```

```python
df.BillAmount.fillna(df.BillAmount.mean(),inplace=True)
```

```python
df
```

```python
#7. Merge the billing dataset with patient dataset on PatientID.
```

```python
a1=df.merge(df,on='PatientID')
```

```python
a1
```

```python
#8. Concatenate an additional DataFrame that contains new patients for the current week (row-wise).
```

```python
df_new=pd.DataFrame({'PatientID':[106,107],'Department':['Cardiology','Neurology'],'Doctor': ['Dr. Smith', 'Dr. Allen'],'BillAmount': [1500, 2000]})
```

```python
a2=pd.concat([a1,df_new],axis=0,ignore_index=True)
a2
```

```python
#9..Concatenate new billing category columns like ['InsuranceCovered', 'FinalAmount'] (column-wise).
```

```python
bill_col=pd.DataFrame({'InsuranceCovered':[1000,1500],'FinalAmount':[1500,2000]})
```

```python
a3=pd.concat([a2.reset_index(drop=True),bill_col.reset_index(drop=True)], axis=1)
```
