```python
# Bar Chart Analysis for Categorical Column
```

```python
categorical_col=df.select_dtypes(include=['object']).columns
```

```python
categorical_col
```

```python
plt.figure(figsize=(15,12))
for i, col in enumerate(categorical_col,1):
    plt.subplot(len(categorical_col)//2+1,2,i)
    df[col].value_counts().plot(kind='bar')
    plt.title(f"Frequency of {col}")
    plt.xlabel(col)
    plt.ylabel("Count")
plt.tight_layout()
plt.show()
    
```

Key Findings :
The business primarily runs on many small, consistent transactions.its averages are shaped by a few large, high-value sales. Median values give a more realistic view of everyday sales patterns. TThe dataset is dominated by small to mid-range transactions, while a few high-value sales and expensive products introduce heavy right skew and outliers. Sales are concentrated in specific days, cities, and product categories. The business is brand-driven, with only a few models/SKUs accounting for most sales. These findings can guide inventory planning, targeted promotions, and city-specific marketing strategies.
